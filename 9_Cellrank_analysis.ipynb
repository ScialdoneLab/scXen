```python
velo_folder='./RNA_velo_processed_data/'
adata_SIGAH5=ad.read_h5ad(velo_folder+'data_SIGAH5.h5ad')
adata_SIGAH12=ad.read_h5ad(velo_folder+'data_SIGAH12.h5ad')
```

# High-level analysis

```python
cr.tl.terminal_states(adata_SIGAH5, cluster_key='seurat_clusters', weight_connectivities=0.2)
cr.pl.terminal_states(adata_SIGAH5,basis='umap2')
```

```python
label='2021_09_24_CellRank_'
# Compute terminal and initial states
cr.tl.terminal_states(adata_SIGAH5, cluster_key='seurat_clusters', weight_connectivities=0.2)
cr.tl.terminal_states(adata_SIGAH12, cluster_key='seurat_clusters', weight_connectivities=0.2)

cr.pl.terminal_states(adata_SIGAH5,basis='umap2',save=label+'term_states_SIGAH5_IVF.pdf')
cr.pl.terminal_states(adata_SIGAH12,basis='umap2',save=label+'term_states_SIGAH12_NT.pdf')
cr.tl.initial_states(adata_SIGAH5, cluster_key='seurat_clusters')
cr.tl.initial_states(adata_SIGAH12, cluster_key='seurat_clusters')
cr.pl.initial_states(adata_SIGAH5, discrete=True,basis='umap2',save=label+'init_states_SIGAH5_IVF.pdf')
cr.pl.initial_states(adata_SIGAH12, discrete=True,basis='umap2',save=label+'init_states_SIGAH12_NT.pdf')
```

```python
# Compute lineages
cr.tl.lineages(adata_SIGAH5)
cr.tl.lineages(adata_SIGAH12)
cr.pl.lineages(adata_SIGAH5, same_plot=False,basis='umap2',save=label+'sep_lineages_SIGAH5_IVF.pdf')
cr.pl.lineages(adata_SIGAH12, same_plot=False,basis='umap2',save=label+'sep_lineages_SIGAH12_NT.pdf')
cr.pl.lineages(adata_SIGAH5, same_plot=True,basis='umap2',save=label+'all_lineages_SIGAH5_IVF.pdf')
cr.pl.lineages(adata_SIGAH12, same_plot=True,basis='umap2',save=label+'all_lineages_SIGAH12_NT.pdf')
```
